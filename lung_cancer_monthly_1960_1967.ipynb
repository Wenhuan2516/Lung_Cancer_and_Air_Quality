### Clean Age

```python
sd['age'].unique()
```

```python
import numpy as np
```

```python
def ageRange(age):
    if age <= 4:
        return 'age 0-24'
    elif age >= 5 and age <= 8:
        return 'age 25-64'
    elif age >= 9 and age <= 11:
        return '65 and older'
    else:
        return np.nan
```

```python
sd['age_range'] = sd['age'].apply(ageRange)
sd.head()
```
